```python
import os
from nltk.tokenize import word_tokenize
```

```python
files = os.listdir('Abstracts')
filePairs = [] #(1,'1.txt')

for file in files:
    num = int(file.replace('.txt',''))
    filePairs.append((num,file))

filePairs.sort() #sorts by number as it is first item in tuple

sortedFiles = []
for i in filePairs:
    sortedFiles.append(i[1]) #appending file name
```

```python
invertedIndex = {}

for file in sortedFiles:
    with open('Abstracts/'+file,'r') as f:
        data = f.read()
    words = word_tokenize(data)

    docID = int(file.replace('.txt',''))
    for word in words:
        if word not in invertedIndex:
            invertedIndex[word] = [docID]
        elif docID not in invertedIndex[word]:
            invertedIndex[word].append(docID)

```

## Inverted Index Created (Now will sort it alphabetically)

```python
import string
invertedIndex = dict(sorted(invertedIndex.items()))
print("Total terms in index: ",len(invertedIndex))
```

```python
for term,docIDs in invertedIndex.items():
    print(f"{term}: {docIDs}")
```

```python
print("Total terms in index: ",len(invertedIndex))
```

```python
from nltk.stem.porter import PorterStemmer

porterStemmer = PorterStemmer()

query = "autoencoders"  
stemmed_query = porterStemmer.stem(query)
print(stemmed_query)
if stemmed_query in invertedIndex:
    print("Query Result:", invertedIndex[stemmed_query])
else:
    print("No results found")
```

# Inverted Index Done. Now gonna store it in a file

```python
with open("22K4036_invertedIndex.txt",'w') as f:
    for term,docIDs in invertedIndex.items():
        f.write(f"{term} -> {docIDs}\n")
        
print("Inverted Index Saved")
```

# Now going to implement a positional index to work with proximity queries

```python
files = os.listdir('Abstracts')
filePairs = [] #(1,'1.txt')

for file in files:
    num = int(file.replace('.txt',''))
    filePairs.append((num,file))

filePairs.sort() #sorts by number as it is first item in tuple

sortedFiles = []
for i in filePairs:
    sortedFiles.append(i[1]) #appending file name
```

### Data Structure: Dictionary within a Dictionary

```python
positionalIndex = {}

for file in sortedFiles:
    docID = int(file.replace('.txt',''))
    with open('Abstracts/'+file,'r') as f:
        data = f.read()
    terms = word_tokenize(data)
    position = 0 #tracking positions of terms
    for term in terms:
        if term not in positionalIndex:
            positionalIndex[term] = {}
        if docID not in positionalIndex[term]:
            positionalIndex[term][docID] = []
        positionalIndex[term][docID].append(position)
        position = position + 1
```

```python
positionalIndex = dict(sorted(positionalIndex.items()))
for term,docIDs in positionalIndex.items():
    print(f"{term}: {docIDs}")
```

```python
print(len(positionalIndex.items()))
```

# Positional Index Created. Now storing to file

```python
with open('22K4036_positionalIndex.txt','w') as f:
    for term,docIDs in positionalIndex.items():
        f.write(f"{term} -> {docIDs}\n")
print("Positional Index Saved in File")
```
